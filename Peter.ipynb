```python
def errorsents(fileid):
    tagsenti=Peter.tagged_sents(fileid, relation = True, speaker = ['CHI'])
    num = len(Peter.sents(fileid))
    taglist=[] 
    taglist.append(tagsenti[:num])
    for tag in taglist:
        for w in pronounlistmin:
            if w in tag:
                taglist.append(tag)
    return taglist
def errorlist(fileid):
    countlist = []
    nnlist = []
    taglist = errorsents(fileid)
    for sent in taglist:
        for ss in sent:
            for nn in ss:
                for w in pronounlistmin:
                    if w in nn:
                        nnlist.append(nn)
    for n in nnlist:
        nlist=list(n)
        del nlist[1]
        a = tuple(i.split("|")[-1] for i in nlist)
        countlist.append(a)
    return countlist
def errornum(fileid):
    countlist = errorlist(fileid)
    errorcountlist=[countlist.count(er) for er in pronounerrorlist]
    return errorcountlist
def errornumlist(fileid):
    countlist = errorlist(fileid)
    errorcountlist=[countlist.count(er) for er in pronounerrorlist]
    return errorcountlist
```

```python
def totalpronoun(fileid):
    freqdist = FreqDist(Peter.words(fileid,speaker= ['CHI']))  
    totalpronoun = [freqdist[i] for i in pronounlist ]
    return totalpronoun
def totalpronounmin(fileid):
    freqdist = FreqDist(Peter.words(fileid,speaker= ['CHI']))  
    totalpronounmin = [freqdist[t] for t in pronounlistmin]
    return totalpronounmin
```
